```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
```

```python
df1 =pd.read_csv('advanced96/my_model_advanced96.csv')
```

```python
df2 =pd.read_csv('advanced96/my_model_advanced96_30.csv')
```

```python
df3 =pd.read_csv('advanced96/my_model_advanced96_60.csv')
```

```python
df4 =pd.read_csv('advanced96/my_model_advanced96_100.csv')
```

```python
df5 =pd.read_csv('advanced96/my_model_advanced96_130.csv')
```

```python
df6 =pd.read_csv('advanced96/my_model_advanced96_150.csv')
```

```python
col = ['loss','accuracy','val_loss','val_accuracy']
```

```python
df = pd.concat([df1[col], df2[col],df3[col],df4[col],df5[col],df6[col]])
```

```python
df= df.reset_index()
```

```python
df
```

```python
plt.plot(df['accuracy'])
plt.plot(df['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

```python
plt.plot(df['loss'])
plt.plot(df['val_loss'])
plt.title('model Loss')
plt.ylabel('Loss')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```
