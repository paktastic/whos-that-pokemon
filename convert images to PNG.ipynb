remove corrupted pictures

```python
dataset_dir = 'output/train'
for foldername in os.listdir(dataset_dir):
    for picture in os.listdir(dataset_dir +'/' + foldername):
        file_path = os.path.join(dataset_dir +'/' + foldername, picture)

        try:
            # Open and load the image using PIL
            img = Image.open(file_path)
            img.load()
            img.close()

            # Image is not corrupted
            #print(f"{filename} is not corrupted.")

        except (IOError, SyntaxError) as e:
            # Image is corrupted
            print(f"{file_path} is corrupted and removed.")
            os.remove(file_path)
```

check for pallette

```python
dataset_dir = 'output/train'
for foldername in os.listdir(dataset_dir):
    for picture in os.listdir(dataset_dir +'/' + foldername):
        file_path = os.path.join(dataset_dir +'/' + foldername, picture)

        # Open the image
        image = Image.open(file_path)

        # Check the image mode
        if image.mode != 'RGBA':
            image.convert("RGBA").save(file_path)
            print(f'{file_path} is now RGBA image.')
```
